```python
import pandas as pd
```

```python
df = pd.read_csv("/home/rmuthy2/Disease_Classification/AP_View.csv")
```

```python
findingsdf = pd.read_csv("/home/rmuthy2/central_data/datasets/internal/EmoryCXRv2/TABLES/EmoryCXR_v2_FindingLabel_10162024.csv")
```

```python
findingsdf.head()
```

```python
findingsdf["Lung Lesion"].value_counts()
```

```python
ll_only = findingsdf[['AccessionNumber_anon', 'Lung Lesion']] 
merged = df.merge(ll_only, on='AccessionNumber_anon')
```

```python
merged
```

```python
merged["Lung Lesion"].value_counts()
```

```python
merged = merged.replace(-1.0, 0.0)
```

```python
merged = merged.fillna(0)
```

```python
merged
```

```python
merged["Lung Lesion"].value_counts()
```

```python
merged
```

```python
positive_cases = merged[merged['Lung Lesion'] == 1]
negative_cases = merged[merged['Lung Lesion']==0]

# Sample same number of negative cases as positive cases
n_positive = len(positive_cases)
negative_sampled = negative_cases.sample(n=n_positive, random_state=42)

# Combine both
balanced_df = pd.concat([positive_cases, negative_sampled], ignore_index=True)

# Shuffle the final dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"Positive cases: {n_positive}")
print(f"Negative cases: {len(negative_sampled)}")
print(f"Total balanced dataset: {len(balanced_df)}")
```

```python
balanced_df.to_csv('llcohort.csv', index=False)
```
